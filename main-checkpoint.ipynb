```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.imdb.com/chart/top/'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) '
                  'AppleWebKit/537.36 (KHTML, like Gecko) '
                  'Chrome/122.0.0.0 Safari/537.36'
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")


movie_items = soup.find_all('li', class_='ipc-metadata-list-summary-item')
list=[]

for index, movie in enumerate(movie_items, start=1):
    
    title_tag = movie.find('h3', class_='ipc-title__text')
    title = title_tag.text.strip() if title_tag else 'No Title'

    
    rating_tag = movie.find('span', class_='ipc-rating-star--rating')
    rating = rating_tag.text.strip() if rating_tag else 'No Rating'

    data={
        "id":index,
        "Movie":title,
        "Rating":rating,
        }
    list.append(data);

for n in list:
    print(n['id'],'-',n['Movie'],'-',n['Rating']);
    
    
    
```

```python
print(list)
```

```python
import pandas as pd

df=pd.DataFrame(list)
df.to_csv('imdb_top_25_movies.csv',index=False)
```

```python
dt=pd.read_csv('imdb_top_25_movies.csv')
print(dt)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.barh(dt['Movie'],dt['Rating'],color='skyblue')
plt.xlabel('IMDB Rating')
plt.title('Top Movies and Their Ratings')
# plt.gca().invert_yaxi()
plt.tight_layout()
plt.show()
```
